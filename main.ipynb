<h3>Parkinson's Disease Detection</h3>

```python
#Importing files
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
```

```python
#Data Collections (loading csv file)
parkinsons_data=pd.read_csv('parkinsons_data.csv')
```

```python
#gruping the data based on the target variable
# Select only numeric columns
data = parkinsons_data.drop(['name'], axis=1)
data.groupby(parkinsons_data['status']).mean()
```

<h4>Data Pre-processing</h4>

```python
#Separating the features & Target
X = parkinsons_data.drop(columns=['name','status'],axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splittng the data to training data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

<h5>Model Training</h5>

Support Vector Machine Model

```python
model= svm.SVC(kernel='linear')
```

```python
#training the SVM model with training data
model.fit(X_train, Y_train)
```

<h5>Model Evaluation</h5>

Accuracy Score

```python
#accuracy score on training data
X_train_predction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_predction)
```

```python
print('Accuracy score of training data: ',training_data_accuracy)
```

```python
#accuracy score on training data
X_test_predction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_predction)
```

```python
print('Accurac Score of test data: ', test_data_accuracy)
```

<h4>Building a Predictive system</h4>

```python
input_data = (197.07600,206.89600,192.05500,0.00289,0.00001,0.00166,0.00168,0.00498,0.01098,0.09700,0.00563,0.00680,0.00802,0.01689,0.00339,26.77500,0.422229,0.741367,-7.348300,0.177551,1.743867,0.085569)


#changing input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

# standarize the data
std_data=scaler.transform(input_data_reshape)

prediction = model.predict(std_data)

print(prediction)

if (prediction[0] == 0):
    print("The person doesn't have Parkinsons Diseases")
else:
    print("The Person have a Parkinsons")
```
